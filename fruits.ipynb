```python

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline

# Following will create a numpy array for an image
img = mpimg.imread('./Training/Apple Braeburn/0_100.jpg')
print(img.shape)
# Display image using matplotlib from a numpy array
plt.imshow(img)
```

```python
import os
train_categories = []
train_samples = []

for i in os.listdir('./Training'):
    train_categories.append(i)
    train_samples.append(len(os.listdir('./Training/' + i))) # number of sample in a categories file

test_categories = []
test_samples = []
    
for i in os.listdir('./Test'):
    test_categories.append(i)
    test_samples.append(len(os.listdir('./Test/' + i)))

print('number of training categories and number of samples : {}, {}'.format(len(train_categories),sum(train_samples)))    
print('number of test categories and number of samples : {}, {}'.format(len(test_categories),sum(test_samples)))
```

```python

index = np.arange(len(train_categories))
plt.bar(index, train_samples,color = 'g')
plt.xlabel('Categories', fontsize=25)
plt.ylabel('No. of samples', fontsize=25)
plt.xticks(index, train_categories, fontsize=20, rotation=90)
plt.title('Category wise training sample distribution', fontsize=35)
plt.show()


index = np.arange(len(test_categories))
plt.bar(index,test_samples,color = 'g')
plt.xlabel('Categories',fontsize = 25)
plt.ylabel('No. of sample', fontsize = 25)
plt.xticks(index,test_categories,fontsize=2, rotation=90)
plt.title('Category wise test sample distribution', fontsize=35)
plt.show()
```

```python
# 
# Convert all the training and test set into a numpy array
# X will be the image array 3 dimensional and Y will be the one hot encoded vector 1x81 dimension
train = []
test = []

for i in os.listdir("./Training"):
    one_hot = np.zeros(shape=[len(train_categories)])
    actual_index = train_categories.index(i)
    one_hot[actual_index] = 1
    for files in os.listdir("./Training/"+ i):
        img_array = mpimg.imread("./Training/"+ i + "/" + files)
        train.append([img_array, one_hot])
    print("Train Category Status: {}/{}".format(actual_index+1, len(train_categories)))
    
for i in os.listdir("./Test"):
    one_hot = np.zeros(shape=[len(test_categories)])
    actual_index = test_categories.index(i)
    one_hot[actual_index] = 1
    for files in os.listdir("./Test/"+ i):
        img_array = mpimg.imread("./Test/"+ i + "/" + files)
        test.append([img_array, one_hot])
    print("Test Category Status: {}/{}".format(actual_index+1, len(test_categories)))
```
